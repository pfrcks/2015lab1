```python
import itertools
def prizedoor_sim(num):
    return np.random.randint(0,3, (num))

def guess_sim(num):
    return np.random.randint(0,3, (num))

def find_goatdoor(prizedoor, guessedoor):
    goat_hash = [1, 2 ,0, 0 , 1]
    result = []
    for p, g in itertools.izip(prizedoor, guessedoor):
        result.append(goat_hash[p+g])
    return result
    
def switch_guess(goatdoors, guesses):
    switch_hash=[0, 2, 1, 0]
    result = []
    for p, g in itertools.izip(goatdoors, guesses):
        result.append(switch_hash[p+g])
    return result

def win_percent(guesses, prizedoor):
    return 100* (guesses == prizedoor).mean()

nsim = 10000
print win_percent(guess_sim(nsim), prizedoor_sim(nsim))
```

```python
guess = guess_sim(nsim)
prize = prizedoor_sim(nsim)
goats = find_goatdoor(prize, guess)
switch = switch_guess(goats, guess)

print win_percent(switch, prize)
```
